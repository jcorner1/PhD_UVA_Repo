```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Generate 1000 values from a uniform distribution
u = np.random.uniform(0, 1, 1000)
```

```python
# Plot histogram of u
plt.hist(u, bins=30, color='blue', alpha=0.7)
plt.title('Histogram of Uniform Distribution (u)')
plt.xlabel('Values')
plt.ylabel('Count')
plt.show()
```

```python
# Plot histogram with density
plt.hist(u, bins=30, density=True, color='green', alpha=0.7)
plt.axhline(y=1, color='r', linestyle='-')
plt.title('Density Histogram of Uniform Distribution (u)')
plt.xlabel('Values')
plt.ylabel('Density')
plt.show()
```

```python
# Calculate the 80% quantile, median, mean, and standard deviation of u
quantile_80 = np.quantile(u, 0.8)
median_u = np.median(u)
mean_u = np.mean(u)
std_dev_u = np.std(u)

print(f"80% Quantile of u: {quantile_80}")
print(f"Median of u: {median_u}")
print(f"Mean of u: {mean_u}")
print(f"Standard Deviation of u: {std_dev_u}")
```

```python
# Passing the values through an inverse Normal CDF
y = stats.norm.ppf(u)
```

```python
# Plot histogram of y
plt.hist(y, bins=30, color='red', alpha=0.7)
plt.title('Histogram of Transformed Distribution (y)')
plt.xlabel('Values')
plt.ylabel('Count')
plt.show()
```

```python
# Plot histogram with density
plt.hist(y, bins=30, density=True, color='purple', alpha=0.7)
x=np.linspace(-4,4,1000)
plt.plot(x, stats.norm.pdf(x),'r-',lw=2)
plt.title('Density Histogram of Transformed Distribution (y)')
plt.xlabel('Values')
plt.ylabel('Density')
plt.show()
```

```python
# Calculate the 80% quantile, median, mean, and standard deviation of y
quantile_80_y = np.quantile(y, 0.8)
median_y = np.median(y)
mean_y = np.mean(y)
std_dev_y = np.std(y)

print(f"80% Quantile of y: {quantile_80_y}")
print(f"Median of y: {median_y}")
print(f"Mean of y: {mean_y}")
print(f"Standard Deviation of y: {std_dev_y}")
```

```python
# Generate a random sample from a standard normal distribution
std_normal_sample = np.random.normal(0, 1, 100000)

# Plot histogram with density
plt.hist(std_normal_sample, bins=50, density=True, color='orange', alpha=0.7)
plt.plot(x, stats.norm.pdf(x),'r-',lw=2)
plt.title('Density Histogram of Standard Normal Distribution')
plt.xlabel('Values')
plt.ylabel('Density')
plt.show()
```

```python
# Calculate the 80% quantile, median, mean, and standard deviation of the standard normal sample
quantile_80_std_normal = np.quantile(std_normal_sample, 0.8)
median_std_normal = np.median(std_normal_sample)
mean_std_normal = np.mean(std_normal_sample)
std_dev_std_normal = np.std(std_normal_sample)

print(f"80% Quantile of Standard Normal Sample: {quantile_80_std_normal}")
print(f"Median of Standard Normal Sample: {median_std_normal}")
print(f"Mean of Standard Normal Sample: {mean_std_normal}")
print(f"Standard Deviation of Standard Normal Sample: {std_dev_std_normal}")
```
