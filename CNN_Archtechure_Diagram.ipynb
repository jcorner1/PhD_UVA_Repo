```python
from pydrawnet.renderers import SeqRenderer, FreeformRenderer
from pydrawnet import layers, operations
import math
import matplotlib.pyplot as plt
```

```python
SR = SeqRenderer()

# conv and avg pool layers
SR.add_layer(layers.Layer2D(3, 30, 49, "Input", color_light=(0.7, 0.7, 0.5)))
SR.add_operation(operations.ArrowOp(''))

SR.add_layer(layers.Layer2D(12, 30, 49, "Conv2D", color_light=(0.7, 0.7, 0.5)))
SR.add_operation(operations.ArrowOp(''))

SR.add_layer(layers.Layer2D(24, 30, 49, "Conv2D", color_light=(0.7, 0.7, 0.5)))
SR.add_operation(operations.ArrowOp(''))

SR.add_layer(layers.Layer2D(24, 10, 17, "Pool 2D", color_light=(0.7, 0.7, 0.5)))
SR.add_operation(operations.ArrowOp(''))

SR.add_layer(layers.Layer2D(48, 10, 17, "Conv2D", color_light=(0.7, 0.7, 0.5)))
SR.add_operation(operations.ArrowOp(''))

SR.add_layer(layers.Layer2D(48, 4, 6, "Pool 2D", color_light=(0.7, 0.7, 0.5)))
SR.add_operation(operations.ArrowOp(''))

#flatten layer
SR.add_layer(layers.BlockLayer(5, 70, 'Flatten'))

#dense layer
SR.add_operation(operations.DenseOp(1, 20, "Dense Layers", line_kwargs={"linewidths": 0.5}))
SR.add_layer(layers.Layer1D(20, 20, "Features", shape_spacing=0.5))

SR.add_operation(operations.DenseOp(20, 10, "", line_kwargs={"linewidths": 0.5}))
SR.add_layer(layers.Layer1D(10, 10, "Features", shape_spacing=0.5))

SR.add_operation(operations.DenseOp(10, 5, "", line_kwargs={"linewidths": 0.5}))
SR.add_layer(layers.Layer1D(5, 5, "Features", shape_spacing=0.5))

SR.add_operation(operations.DenseOp(5, 1, "", line_kwargs={"linewidths": 0.5}))
SR.add_layer(layers.Layer1D(1, 1, "Features", shape_spacing=0.5))

SR.make_figure((20, 7))
SR.render(75, 150, text_y_offset=20)
```
